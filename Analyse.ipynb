```python
import sqlite3, pandas , matplotlib.pyplot as plt
```

```python
conn = sqlite3.connect("./steam_info.db")
```

```python
sql = """SELECT steamid, 
case name when 'Scout.accum.iPlayTime' then 'Scout'
when 'Soldier.accum.iPlayTime' then 'Soldier'
when 'Spy.accum.iPlayTime' then 'Spy'
when 'Pyro.accum.iPlayTime' then 'Pyro'
when 'Medic.accum.iPlayTime' then 'Medic'
when 'Heavy.accum.iPlayTime' then 'Heavy'
when 'Demoman.accum.iPlayTime' then 'Demoman'
when 'Engineer.accum.iPlayTime' then 'Engineer'
when 'Sniper.accum.iPlayTime' then 'Sniper' end as name, value / 60.0 / 60.0 as hours
FROM TF2stats 
INNER JOIN steamids_checked ON steamids_checked.id = TF2stats.steamid_id"""

data = pandas.read_sql(sql, conn)
plt.bar(data.name, data.hours)
plt.title("TF2 class hours")
plt.show()
```
